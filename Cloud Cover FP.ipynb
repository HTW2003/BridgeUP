```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
cloud_frac = open("CloudFraction1_2002.csv", 'r') 
cloud_reader = csv.reader(cloud_frac, delimiter=',')
cloud_lists = list(cloud_reader)
```

```python
print(len(cloud_lists[0])) #items per row
```

```python
#print(cloud_lists)
```

```python
print(len(cloud_lists)) #rows
```

```python
width = len(cloud_lists[0])
height = len(cloud_lists)
```

```python
cloud_grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
cloud_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(cloud_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            cloud_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 0: #change conditionals
            cloud_grid[r, c] = [0, 0, 255]  #[R, G, B]
        elif float(value) <= 0.1:
            cloud_grid[r, c] = [79, 131, 249]
        elif float(value) <= 0.2:
            cloud_grid[r, c] = [109, 153, 255]
        elif float(value) <= 0.3:
            cloud_grid[r, c] = [135, 171, 255]
        elif float(value) <= 0.4:
            cloud_grid[r, c] = [149, 179, 249]
        elif float(value) <= 0.5:
            cloud_grid[r, c] = [164, 189, 249]
        elif float(value) <= 0.6:
            cloud_grid[r, c] = [182, 201, 249] #add more elifs
        elif float(value) <= 0.7:
            cloud_grid[r, c] = [195, 209, 244]
        elif float(value) <= 0.8:
            cloud_grid[r, c] = [214, 224, 249]
        elif float(value) <= 0.9:
            cloud_grid[r, c] = [234, 239, 252]
        elif float(value) <= 1.0:
            cloud_grid[r, c] = [255, 255, 255]
        else:
            cloud_grid[r, c] = [0, 0, 0]
```

```python
print(min(cloud_values))
print(max(cloud_values))
```

```python
#2002 Cloud Cover
plt.title("Cloud Cover 2002")
plt.imshow(cloud_grid)
plt.axis('off')
plt.savefig("CloudFraction1_2002.png",dpi = 1000)
plt.show()
cloud_frac = open("CloudFraction1_2002.csv") 
cloud_reader = csv.reader(cloud_frac, delimiter=',')
cloud_lists = list(cloud_reader)
```

```python
NY = pd.read_csv("Annual_NY_CloudCover_csv.csv")
```

```python
NY.head()
```

```python
year = NY["Year"]
total = NY["Total Cloud Cover"]
temp = NY["Temperature �"]
rad = NY["Daily Shortwave Radiation"]
```

```python
print(year.head())
print(total.head())
print(temp.head())
print(rad.head())
```

```python
year.max()
```

```python
#Total cloud cover and daily shortwave radiation from 1985-2018
plt.plot(year, total, color = "blue") #Total cloud cover
plt.legend()
plt.twinx() #2 y-axes
plt.plot(year, rad, color = "red") #Daily shortwave radiation
plt.title("Annual NY Cloud Cover")
plt.legend(loc = "upper left") #Position of the legend
plt.xlim(1985, 2018)
plt.xticks(range(1985, 2018, 10))
plt.savefig("NY annual CloudCover cc + rad")
plt.show()
```

```python
#Temperature from 1985-2018
plt.plot(year, temp) #Temperature
plt.title("Annual NY Cloud Cover")
plt.legend()
plt.xlim(1985, 2018)
plt.xticks(range(1985, 2018, 10))
plt.savefig("NY annual CloudCover temp")
plt.show()
```
