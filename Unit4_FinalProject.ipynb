### Let's add some color (and functions)
Next, you'll need to color code your points based on one of the characteristics in the data frame. 

First, create a function that takes the name of a column and a number of colors and returns list of intermediate values based on the number of colors you specify (these values will be the boundaries for conditionals later on).

```python
lon = df["LONGITUDE"]
lat = df["LATITUDE"]
deaths = df["DEATHS"]
mag = df["EQ_PRIMARY"]
cm = plt.cm.get_cmap("Greens")
image = plt.imread("earth.jpg")
fig, ax = plt.subplots()
ax.imshow(image, extent = [-180,180,-90,90])
plt.scatter(lon, lat, color = "orange")
plt.scatter(lon, lat, c = deaths, s=mag*10, cmap=cm, norm=mpl.LogNorm())
plt.xlim(-180, -50)
plt.ylim(0, 90)
plt.title("North America")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.savefig("Kill me now, please")
plt.show()
```

```python
#Our function
def yeet(lonme, latme):
    lon = df[lonme]
    lat = df[latme]
    #Background map
    image = plt.imread("plates.png")
    fig, ax = plt.subplots()
    ax.imshow(image, extent = [-180,180,-90,90])
    #Scatter plot
    plt.scatter(lon, lat, color = "orange")
    plt.scatter(lon, lat, c = deaths, s=mag*10, cmap=cm, norm=mpl.LogNorm())
    plt.xlim(-180, -50)
    plt.ylim(0, 90)
#Using the function
yeet("LONGITUDE", "LATITUDE")
plt.title("North America")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.savefig("I can make functions!")
plt.show()
```
